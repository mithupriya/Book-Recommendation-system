```python
def clean_book_title(df):
    document=list(df['book_title'] . values)
    book_title=[re.findall(r'name=(:,)',sentence) for sentence in document]
    book_title=[' '.join(word) for word in book_title]
    return(book_title)
```

```python
def clean_genres(df):
    document=list(df['genres'] . values)
    genres=[re.findall(r'name=(|.)',sentence) for sentence in document]
    genres=[' '.join(word) for word in genres]
    return(genres)
```

```python
def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:30]
    book_indices = [i[0] for i in sim_scores]
    return titles.iloc[book_indices]
```
